```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread(r'C:\Users\abhim\Desktop\dl\computer_vision\open_cv\face_detection\istockphoto-1368965646-1024x1024.jpg')
```

```python
img_conv=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
face_cascade=cv2.CascadeClassifier(r'C:\Users\abhim\Desktop\dl\computer_vision\open_cv\face_detection\haarcascade_frontalface_default.xml')
eye_cascade=cv2.CascadeClassifier(r'C:\Users\abhim\Desktop\dl\computer_vision\open_cv\face_detection\haarcascade_eye.xml')
```

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
faces=face_cascade.detectMultiScale(gray_img,scaleFactor=1.1,minNeighbors=1)
```

```python
for x,y,w,h in faces:
    cv2.rectangle(img_conv,(x,y),(x+w,y+h),(0,255,0),2)
    eyes=eye_cascade.detectMultiScale(gray_img,scaleFactor=1.1,minNeighbors=4)
    for ex,ey,ew,eh in eyes:
        cv2.rectangle(img_conv,(ex,ey),(ex+ew,ey+eh),(255,0,0),1)
```
